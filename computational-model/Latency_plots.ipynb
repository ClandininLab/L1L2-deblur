# Import packages and read pickle file

```python
import pandas as pd
import pickle
import copy
from matplotlib import pyplot as plt
from matplotlib import axes as ax
from scipy import stats as st
import seaborn
```

```python
df = pd.read_pickle("data_shared.pickle")
```

# Fig 4 (vertical)

```python
lat_vert = df['latency']['vertical']
```

```python
lat_vert_300_models = copy.deepcopy(lat_vert)
desired_order_list = ['modelnofeedbackSpeed300', 'modelfittedSpeed300']
lat_vert_300_models = {k: lat_vert_300_models[k] for k in desired_order_list}
```

```python
labels, data = lat_vert_300_models.keys(), lat_vert_300_models.values()
plt.violinplot(data, vert=False, showextrema=False, quantiles = [[0.25, 0.5, 0.75],[0.25, 0.5, 0.75]])
labels = ['Recurrent model', 'Non-recurrent model']
plt.yticks(range(1,len(labels)+1), labels)
plt.xlabel('Temporal offset (ms)')
plt.title('Vertical motion')
plt.xlim(-175,175)
plt.savefig('vert.svg')
plt.show()
```

```python
st.wilcoxon(lat_vert['modelfittedSpeed300'], lat_vert['modelnofeedbackSpeed300'])
```

# Fig 4 (horizontal)

```python
lat_horiz = df['latency']['horizontal']
```

```python
lat_horiz_300_models = copy.deepcopy(lat_horiz)
desired_order_list = ['modelnofeedbackSpeed300', 'modelfittedSpeed300']
lat_horiz_300_models = {k: lat_horiz_300_models[k] for k in desired_order_list}
```

```python
labels, data = lat_horiz_300_models.keys(), lat_horiz_300_models.values()
plt.violinplot(data, vert=False, showextrema=False, showmedians=True, quantiles = [[0.25, 0.5, 0.75],[0.25, 0.5, 0.75]])
labels = ['Recurrent model', 'Non-recurrent model']
plt.yticks(range(1,len(labels)+1), labels)
plt.xlabel('Temporal offset (ms)')
plt.title('Horizontal motion')
plt.xlim(-175,175)
plt.savefig('horiz.svg')
plt.show()
```

```python
st.wilcoxon(lat_horiz_300['modelfittedSpeed300'], lat_horiz_300['modelnofeedbackSpeed300'])
```
